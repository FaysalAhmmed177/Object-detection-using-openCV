```python
#Import library
import cv2
```

```python
#Threshold to detect object
thres = 0.45
```

```python
#live video Capture by the default camera
cap = cv2.VideoCapture(0)
cap.set(3,1280)
cap.set(4,720)
cap.set(10,70)
```

```python
#img = cv2.imread("final thesis/airplane1.jpg")
```

```python
#define the object class
classNames= []
classFile = 'data/object.name'
with open(classFile,'rt') as f:
    classNames = f.read().rstrip('\n').split('\n')
```

```python
#load the configuration file and weight file
configPath = 'data/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath = 'data/frozen_inference_graph.pb'
```

```python
#apply model
net = cv2.dnn_DetectionModel(weightsPath,configPath)
net.setInputSize(320,320)
net.setInputScale(1.0/ 127.5)
net.setInputMean((127.5, 127.5, 127.5))
net.setInputSwapRB(True)
```

```python
while True:
    
    #Capture frame-by-frame
    success,img = cap.read()
    classIds, confs, bbox = net.detect(img,confThreshold=thres)
    print(classIds,bbox)

    if len(classIds) != 0:
        for classId, confidence,box in zip(classIds.flatten(),confs.flatten(),bbox):
            #showing the object detection rectangle area
            cv2.rectangle(img,box,color=(0,255,0),thickness=2)
            #display the object name
            cv2.putText(img,classNames[classId-1].upper(),(box[0]+10,box[1]+30),
                        cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
            #Showing the accuracy
            cv2.putText(img,str(round(confidence*100,2)),(box[0]+200,box[1]+30),
                        cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
    
    #Display the resulting frame
    cv2.imshow("Final Result",img)
    
     #This commands lets us quit with the 'q' button on a keyboard.
    #Simply pressing X on the window won't work
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
#When everything is done,release the capture       
cap.release()
cv2.destroyAllWindows()
```
